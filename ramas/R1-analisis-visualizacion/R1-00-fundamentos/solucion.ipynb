```python
# ── Preparación del entorno (ejecuta esta celda primero) ──────────────────────
import os, urllib.request
import pandas as pd
import matplotlib.pyplot as plt

# "En vivo o caché": usa el archivo local; si falta (ej. Colab), lo baja del repo.
CSV = "compras_ml.csv"
if not os.path.exists(CSV):
    urllib.request.urlretrieve(f"https://raw.githubusercontent.com/formacionpublica-bootcamp/bootcamp/main/data/compras_ml.csv", CSV)

df = pd.read_csv(CSV)
print("Datos cargados:", df.shape, "filas x columnas")
df.head()
```

### Datos cargados arriba. Empecemos por lo básico de Python.

## 1. Variables: guardar para reutilizar

Una **variable** guarda un valor con nombre. Tomemos los montos de compras reales como una lista y calculemos el total y el promedio sin pandas.

### ✍️ Ejercicio 1 — Total y promedio de los montos

```python
montos = df["monto_total"].tolist()
total = sum(montos)
promedio = total / len(montos)
print("total:", round(total), "| promedio:", round(promedio))
```

```python
# ── Celda de chequeo · Ejercicio 1 ──────────────────────────────────────────
try:
    assert abs(total - df["monto_total"].sum()) < 1e-6
    assert abs(promedio - df["monto_total"].mean()) < 1e-6
    print("✅ Ejercicio 1: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", 'sum(montos) y total / len(montos).')
    print("   Detalle:", e)
```

## 2. Listas y comprensiones: filtrar

Una **comprensión de lista** filtra en una línea. Contemos las compras caras.

### ✍️ Ejercicio 2 — Cuenta los montos sobre $500.000

```python
caros = [m for m in montos if m > 500_000]
n_caros = len(caros)
print("compras > $500.000:", n_caros)
```

```python
# ── Celda de chequeo · Ejercicio 2 ──────────────────────────────────────────
try:
    assert n_caros == int((df["monto_total"] > 500_000).sum())
    print("✅ Ejercicio 2: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", '[m for m in montos if m > 500_000] y luego len(...).')
    print("   Detalle:", e)
```

## 3. Diccionarios y bucles: acumular por clave

Un **diccionario** asocia clave→valor. Con un **bucle** sumamos el gasto por región.

### ✍️ Ejercicio 3 — Gasto total por región con un bucle

```python
gasto_region = {}
for region, monto in zip(df["region_comprador"], df["monto_total"]):
    gasto_region[region] = gasto_region.get(region, 0) + monto
print(len(gasto_region), "regiones")
```

```python
# ── Celda de chequeo · Ejercicio 3 ──────────────────────────────────────────
try:
    assert len(gasto_region) == df["region_comprador"].nunique()
    assert abs(sum(gasto_region.values()) - df["monto_total"].sum()) < 1.0
    print("✅ Ejercicio 3: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", 'gasto_region[region] = gasto_region.get(region, 0) + monto.')
    print("   Detalle:", e)
```

## 4. Condicionales: clasificar en tramos

Con **if/elif/else** dentro de un bucle clasificamos cada compra por tamaño de monto.

### ✍️ Ejercicio 4 — Cuenta las compras por tramo

```python
tramos = {"bajo": 0, "medio": 0, "alto": 0}
for m in montos:
    if m < 100_000:
        tramos["bajo"] += 1
    elif m < 1_000_000:
        tramos["medio"] += 1
    else:
        tramos["alto"] += 1
print(tramos)
```

```python
# ── Celda de chequeo · Ejercicio 4 ──────────────────────────────────────────
try:
    assert sum(tramos.values()) == len(montos)
    assert tramos["bajo"] + tramos["medio"] + tramos["alto"] == len(montos)
    print("✅ Ejercicio 4: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", 'if m < 100_000 / elif m < 1_000_000 / else, sumando 1 al tramo correspondiente.')
    print("   Detalle:", e)
```

```python
# (ilustración) Distribución de compras por tramo de monto
plt.figure(figsize=(5, 4))
plt.bar(tramos.keys(), tramos.values(), color="#4240e5")
plt.ylabel("Nº de compras"); plt.title("Compras por tramo de monto")
plt.tight_layout(); plt.show()
```
