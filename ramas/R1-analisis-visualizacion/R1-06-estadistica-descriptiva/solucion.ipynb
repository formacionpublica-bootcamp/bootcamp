## 1. Por qué un solo número no basta
Imagina que te preguntan: *"¿cuántos habitantes tiene una comuna típica?"*. Podrías responder con **el promedio**. Pero el promedio puede engañar: si en la muestra hay una comuna gigante (Puente Alto) y muchas chicas, el promedio se **infla** hacia arriba y deja de representar a la comuna "típica".

Describir bien un conjunto de datos significa mirar **varias medidas a la vez**:
- **Centro:** ¿alrededor de qué valor se agrupan los datos? (promedio, mediana, moda)
- **Dispersión:** ¿qué tan juntos o separados están? (rango, desviación estándar)
- **Posición:** ¿cómo se reparten? (cuartiles, percentiles)

Veámoslas una por una sobre la población de las comunas.


## 2. Medidas de centro: media, mediana y moda
- **Media (promedio):** suma de todos los valores dividida por la cantidad. En pandas: `.mean()`. Es la más conocida, pero **sensible a los extremos**: un valor muy grande la arrastra hacia arriba.
- **Mediana:** el valor **del medio** cuando ordenas los datos (la mitad está por debajo, la mitad por encima). En pandas: `.median()`. Es **robusta**: un dato extremo casi no la mueve.
- **Moda:** el valor que **más se repite**. En pandas: `.mode()`. Útil en datos con categorías o valores repetidos; en datos donde cada valor es único (como acá, donde cada comuna tiene una población distinta) no aporta mucho.

**La clave:** cuando los datos están **sesgados** (muchos chicos y pocos muy grandes), la media y la mediana se separan. Comparar ambas te dice si hay extremos tirando del promedio.


```python
media = df["poblacion"].mean()
mediana = df["poblacion"].median()

print(f"Media (promedio): {media:,.0f} habitantes")
print(f"Mediana:          {mediana:,.0f} habitantes")
print()
print("La media es bastante MAYOR que la mediana: las comunas grandes")
print("(Puente Alto, Maipú...) arrastran el promedio hacia arriba.")
```

## 3. Medidas de dispersión: rango y desviación estándar
Dos grupos pueden tener el **mismo promedio** y ser muy distintos. Ejemplo: el promedio de [50, 50, 50] y el de [10, 50, 90] es 50 en ambos, pero el segundo está mucho más **disperso**. Las medidas de dispersión capturan eso:

- **Mínimo y máximo:** los valores extremos. `.min()`, `.max()`.
- **Rango:** la diferencia entre el máximo y el mínimo (`max - min`). Da una idea rápida de la amplitud.
- **Desviación estándar:** mide, en promedio, **cuánto se alejan los datos de la media**. En pandas: `.std()`. Está en las **mismas unidades** que los datos (acá, habitantes): mientras más grande, más dispersos. (Su versión al cuadrado es la *varianza*, menos intuitiva porque queda en "habitantes al cuadrado".)


```python
minimo = df["poblacion"].min()
maximo = df["poblacion"].max()
rango = maximo - minimo
desviacion = df["poblacion"].std()

print(f"Mínimo:               {minimo:,.0f}")
print(f"Máximo:               {maximo:,.0f}")
print(f"Rango (máx - mín):    {rango:,.0f}")
print(f"Desviación estándar:  {desviacion:,.0f} habitantes")
```

## 4. Medidas de posición: cuartiles y percentiles
Los **percentiles** parten los datos ordenados en 100 partes; los **cuartiles**, en 4:

- **Q1 (percentil 25):** el 25% de las comunas está por debajo de este valor.
- **Q2 (percentil 50):** es la **mediana**.
- **Q3 (percentil 75):** el 75% está por debajo.

En pandas se obtienen con `.quantile(0.25)`, `.quantile(0.5)`, `.quantile(0.75)`. Sirven para responder preguntas como *"¿cuál es el tamaño bajo el cual está la cuarta parte más pequeña de las comunas?"*.


```python
q1 = df["poblacion"].quantile(0.25)
q2 = df["poblacion"].quantile(0.50)   # = mediana
q3 = df["poblacion"].quantile(0.75)

print(f"Q1 (25%): {q1:,.0f}")
print(f"Q2 (50%, mediana): {q2:,.0f}")
print(f"Q3 (75%): {q3:,.0f}")
```

## 5. `describe()`: todo de una vez
pandas resume las medidas más importantes de una columna numérica con un solo método: `.describe()`. Devuelve el conteo, la media, la desviación estándar, el mínimo, los tres cuartiles y el máximo. Es lo primero que conviene mirar al explorar una columna numérica.


```python
df["poblacion"].describe()
```

## 6. La gran lección: cuántas comunas están **bajo** el promedio
Aquí se ve por qué el promedio engaña con datos sesgados. Si el promedio fuera un buen "valor típico", esperarías que más o menos la mitad de las comunas estén por encima y la otra mitad por debajo. Pero con datos sesgados hacia arriba, **la mayoría queda por debajo del promedio**: unas pocas comunas enormes lo levantan.

Contémoslo: ¿cuántas de las 30 comunas tienen menos población que el promedio?


```python
media = df["poblacion"].mean()
bajo_promedio = (df["poblacion"] < media).sum()

print(f"Promedio: {media:,.0f} habitantes")
print(f"Comunas BAJO el promedio: {bajo_promedio} de {len(df)}")
print()
print("Moraleja: con datos sesgados, reportar solo el promedio da una")
print("impresión equivocada de la comuna 'típica'. La mediana es más fiel.")
```

## ⚠️ Errores típicos de principiante
- **Reportar solo el promedio.** Con datos sesgados o con extremos, el promedio engaña. Acompáñalo siempre de la mediana (y mira la dispersión).
- **Confundir media y mediana.** La media es el promedio aritmético; la mediana es el valor del medio. No son lo mismo, sobre todo con extremos.
- **Olvidar las unidades de la desviación estándar.** La desviación está en las mismas unidades que los datos; la varianza está al cuadrado (poco intuitiva).
- **Comparar grupos solo por la media.** Dos grupos con igual promedio pueden tener dispersiones muy distintas. Mira también el rango y la desviación.
- **Usar la moda en datos continuos.** Si cada valor es casi único (como poblaciones exactas), la moda no aporta.
- **Ignorar los faltantes.** `.mean()` y compañía ignoran los `NaN` por defecto; conviene saber cuántos faltantes hay (lo viste en M4) antes de interpretar.

## Ejercicio 05 — Interpretar el sesgo (media vs mediana)

Hasta ahora calculaste media y mediana por separado. Ahora vas a **interpretar** lo que su comparacion significa.

Una forma sencilla de medir el sesgo es la **razon entre media y mediana**: divide la media por la mediana y guardala en `razon`.

- Si `razon` es **cercana a 1**, media y mediana coinciden: la distribucion es bastante **simetrica**.
- Si `razon` es **claramente mayor que 1**, la media esta por encima de la mediana: la distribucion esta **sesgada hacia la derecha** (unos pocos valores muy grandes inflan el promedio).
- Si `razon` es **claramente menor que 1**, ocurre lo contrario: sesgo hacia la izquierda.

Calcula `razon` y luego, segun el valor que obtengas, **elige la interpretacion correcta** guardando la letra en `conclusion` (por ejemplo, `conclusion = "B"`):

- **A.** La razon es cercana a 1, asi que la poblacion de las comunas es bastante simetrica y el promedio describe bien a la comuna tipica.
- **B.** La razon es bastante mayor que 1, asi que la distribucion esta sesgada hacia la derecha: el promedio se infla por unas pocas comunas enormes y la **mediana** representa mejor a la comuna tipica.
- **C.** La razon es menor que 1, asi que hay muchas comunas grandes y pocas chicas tirando del promedio hacia abajo.

*(Opcional, no se corrige): en la variable de texto `reflexion` escribe en una frase que numero le reportarias a una autoridad que pregunta "¿de que tamano es una comuna tipica?" y por que.)*

Pista: `razon = df["poblacion"].mean() / df["poblacion"].median()`.
