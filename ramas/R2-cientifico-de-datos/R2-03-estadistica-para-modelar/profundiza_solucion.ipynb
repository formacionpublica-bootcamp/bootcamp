```python
import pandas as pd, numpy as np, os, urllib.request
from scipy import stats
import statsmodels.api as sm
CSV = "compras_ml.csv"
if not os.path.exists(CSV):
    urllib.request.urlretrieve("https://raw.githubusercontent.com/formacionpublica-bootcamp/bootcamp/main/data/compras_ml.csv", CSV)
df = pd.read_csv(CSV)
print("Datos:", df.shape)
```

## 1. Intervalo por *bootstrap* (sin fórmulas)

Cuando la distribución es rara, el **bootstrap** estima el intervalo remuestreando los datos. Sin supuestos: solo computación.

### ✍️ Ejercicio 1 — Construye un IC 95% por bootstrap para la media

```python
x = df["monto_total"].values
rng = np.random.default_rng(42)
medias = [rng.choice(x, size=len(x), replace=True).mean() for _ in range(300)]
ic_low, ic_high = np.percentile(medias, [2.5, 97.5])
print("IC95% bootstrap: [", round(ic_low), ",", round(ic_high), "]")
```

```python
# ── Celda de chequeo · Ejercicio 1 ──────────────────────────────────────────
try:
    assert ic_low < df["monto_total"].mean() < ic_high
    print("✅ Ejercicio 1: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", 'Lista por comprensión de 300 rng.choice(x, size=len(x), replace=True).mean().')
    print("   Detalle:", e)
```

## 2. Alternativa robusta: Mann–Whitney

El t-test asume normalidad. Con colas largas, el test no paramétrico de **Mann–Whitney** es más seguro.

### ✍️ Ejercicio 2 — Compara Grande vs Micro sin asumir normalidad

```python
grande = df[df["tamano_proveedor"] == "Grande"]["monto_total"]
micro  = df[df["tamano_proveedor"] == "Micro"]["monto_total"]
u, p = stats.mannwhitneyu(grande, micro, alternative="two-sided")
print("U:", round(u, 1), "| p-valor:", p)
```

```python
# ── Celda de chequeo · Ejercicio 2 ──────────────────────────────────────────
try:
    assert 0.0 <= p <= 1.0
    print("✅ Ejercicio 2: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", "stats.mannwhitneyu(grande, micro, alternative='two-sided').")
    print("   Detalle:", e)
```

## 3. Una primera regresión (statsmodels)

La inferencia se conecta con el modelado: una **regresión lineal** estima cuánto sube el monto por unidad de cantidad, con su significancia.

### ✍️ Ejercicio 3 — Ajusta monto ~ cantidad con OLS

```python
X = sm.add_constant(df["cantidad"])
modelo = sm.OLS(df["monto_total"], X).fit()
r2 = modelo.rsquared
print("R²:", round(r2, 4), "| coef cantidad:", round(modelo.params["cantidad"], 2))
```

```python
# ── Celda de chequeo · Ejercicio 3 ──────────────────────────────────────────
try:
    assert 0.0 <= r2 <= 1.0
    print("✅ Ejercicio 3: ¡correcto!")
except Exception as e:
    print("❌ Aún no. Revisa tu respuesta y vuelve a intentarlo.")
    print("   Pista:", 'sm.OLS(y, X).fit(); el atributo .rsquared es el R².')
    print("   Detalle:", e)
```
