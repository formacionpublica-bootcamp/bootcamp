## 1. ¿Qué es una API? ¿Y qué es JSON?

Una **API** es una *ventanilla de datos*: una dirección de internet (una URL) a la que tu programa le
**pide** información y ella te la **entrega**, lista para usar.

> 🧠 **Analogía.** Piensa en el **mesón de una oficina pública**: tú llegas con una solicitud
> clara ("quiero el valor del dólar de hoy"), y la persona del mesón te devuelve exactamente ese
> dato. No entras a la bodega a buscarlo: pides por la ventanilla y recibes. Una API es esa
> ventanilla, pero para datos.

Lo que la API te devuelve viene en **JSON**, un formato de texto que —por suerte— se traduce
**exactamente** a lo que ya conoces de M1: **diccionarios** (`clave: valor`) y **listas**. Cuando
Python lee un JSON, obtienes diccionarios y listas normales.

```python
import requests
respuesta = requests.get("https://mindicador.cl/api")   # pedimos por la ventanilla
datos = respuesta.json()                                 # lo convertimos a diccionario
```

## 2. Navegar el JSON: diccionarios dentro de diccionarios

La respuesta de `mindicador.cl` es un diccionario donde cada indicador es, a su vez, **otro
diccionario**. Por ejemplo, `datos["dolar"]` se ve así:

```python
{"codigo": "dolar", "nombre": "Dolar observado", "unidad_medida": "Pesos",
 "fecha": "2026-06-19T04:00:00.000Z", "valor": 897.19}
```

Para llegar al número, encadenas las claves: `datos["dolar"]["valor"]`. Es como abrir una carpeta
("dolar") y dentro tomar el papel que dice "valor".

> ⚠️ **Errores típicos**
> - Escribir mal la clave (`"Dolar"` en vez de `"dolar"`): el JSON distingue mayúsculas.
> - Olvidar que `datos["dolar"]` todavía es un diccionario, no el número. Hay que dar el segundo paso.

## 3. Cuando el JSON trae una lista

Muchas APIs entregan **series**: una lista de valores en el tiempo. Si pides la URL del dólar con
`/dolar`, la respuesta trae una clave `"serie"` que es una **lista de diccionarios**, uno por día:

```python
{"codigo": "dolar", "nombre": "Dolar observado",
 "serie": [ {"fecha": "...06-19...", "valor": 897.19},
            {"fecha": "...06-18...", "valor": 883.46}, ... ]}
```

Sobre una lista ya sabes trabajar (M1): `len(serie)` cuenta los puntos, `serie[0]` toma el primero
(el más reciente).

antesala de una tabla.


### Mini-glosario
- **API:** ventanilla de datos en internet a la que tu programa le pide información.
- **URL (endpoint):** la dirección exacta que consultas (`https://mindicador.cl/api/dolar`).
- **JSON:** formato de texto que Python convierte en diccionarios y listas.
- **`requests.get(url).json()`:** pedir a la API y convertir la respuesta en datos de Python.
- **Caché:** guardar un dato ya pedido para no volver a molestar a la API.
