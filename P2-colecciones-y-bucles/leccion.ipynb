## 1. Listas: El archivador vertical numerado
Una **lista** es una estructura de datos que te permite guardar varios valores en un orden específico. Se define abriendo corchetes `[ ]` y separando los elementos con comas.

**La analogía del archivador vertical:**
Imagina un archivador con varios cajones ordenados uno tras otro. Cada cajón guarda un expediente. Para poder buscar un expediente, necesitas saber el número de cajón donde está guardado.

### Reglas clave de las listas en Python:
1.  **La Planta Baja (Piso 0):** En programación, la primera posición de una lista (su índice) no es el 1, sino el **0** (como la planta baja o el nivel de la calle). La segunda posición es el 1, la tercera es el 2, y así sucesivamente.
2.  **Buscar desde el fondo:** Puedes usar índices negativos para acceder a los elementos desde el final. Por ejemplo, `lista[-1]` te entrega el último elemento directo, sin importar el largo de la lista.
3.  **Largo de la lista:** La instrucción `len(nombre_lista)` cuenta cuántos elementos hay dentro del archivador.
4.  **Archivar al final:** La instrucción `.append(valor)` agrega un elemento nuevo al final de la lista.

```python
# Creamos una lista con los montos de gasto de 2021, 2022 y 2023
montos = [9835060128510, 11494882965920, 12483986700171]

print(montos[0])     # Imprime el primer monto (Piso 0): 9835060128510
print(montos[-1])    # Imprime el último monto (2023): 12483986700171
print(len(montos))   # Imprime la cantidad de elementos: 3

# Archivamos un nuevo monto (2024) al final
montos.append(15601131083597)
print(len(montos))   # Ahora muestra 4
```

## 2. Tuplas: Expedientes sellados
Una **tupla** es similar a una lista porque guarda elementos ordenados, pero se escribe utilizando paréntesis `( )` en lugar de corchetes. 

**La analogía de la ley promulgada:**
A diferencia de las listas, una tupla **no puede ser modificada** después de haber sido creada (es inmutable). Es como la fecha en que se promulga una ley en el Diario Oficial: es un registro fijo (día, mes, año) que no debería alterarse por error durante la ejecución del programa. 

Si un dato no debe cambiar bajo ninguna circunstancia, se guarda como tupla.

```python
# Definimos un rango de años fijo
anio_rango = (2021, 2025)
print(anio_rango[0])  # Muestra 2021
print(anio_rango[1])  # Muestra 2025

# Si intentamos modificarlo descomentando la línea de abajo, dará un error:
# anio_rango[0] = 2020
```

## 3. Diccionarios: La Ficha de Licitación (Datos con etiqueta)
Un **diccionario** es una estructura que guarda datos relacionando una **clave** (etiqueta) con un **valor**. Se escribe utilizando llaves `{ }`.

**La analogía del formulario o ficha técnica:**
Imagina la ficha de una licitación en MercadoPúblico. En lugar de aprenderte que el presupuesto está en el "tercer cajón", la ficha tiene etiquetas impresas: `"organismo"`, `"monto"`, `"anio"`. Para leer un dato, solo debes apuntar a la etiqueta.

Es la estructura perfecta para representar **un registro** individual (una fila de una tabla).

```python
# Creamos la ficha de un año de compras
compras_2025 = {
    "anio": 2025,
    "monto": 18430546563769,
    "ordenes": 1867076,
    "organismos": 1165,
}

# Para acceder al dato, usamos corchetes con el nombre de la etiqueta (clave) como texto
print(compras_2025["monto"])       # Imprime 18430546563769
print(compras_2025["organismos"])  # Imprime 1165
```

## 4. Bucle `for`: El revisor de expedientes
El bucle `for` sirve para **repetir un bloque de código** para cada uno de los elementos de una colección.

**La analogía del revisor de expedientes:**
Imagina que tienes una pila de 10 carpetas sobre tu escritorio. Para revisarlas, realizas el mismo proceso con cada una:
1. Tomas la primera carpeta.
2. Realizas el trámite (revisar monto, estampar firma, etc. — esto es el bloque de código indentado).
3. La dejas a un lado y tomas la siguiente.
Repites esto **para cada** (`for`) carpeta de la pila hasta que no queda ninguna.

### El patrón acumulador (Contador de presupuesto)
Un uso clásico del bucle `for` es acumular valores. Imagina que tienes una caja registradora en cero (`total = 0`). A medida que revisas cada carpeta de gasto, abres la carpeta, lees el monto y lo sumas a la caja. Al terminar la pila, el total guardará la suma de todo.

```python
montos = [9835060128510, 11494882965920, 12483986700171]

# Ejemplo 1: Recorrer e imprimir cada monto por separado
for m in montos:
    print(f"Revisando monto: {m} CLP")

print("------------------")

# Ejemplo 2: Sumar acumulando (Patrón acumulador)
total_gasto = 0  # Nuestra caja totalizadora parte en 0
for m in montos:
    total_gasto = total_gasto + m  # Sumamos el monto actual al total acumulado

print(f"Gasto total acumulado: {total_gasto} CLP")
```

## 5. Lista de diccionarios: Una planilla de Excel
En la práctica de la ciencia de datos, los registros vienen organizados en tablas. En Python, la estructura nativa para representar una **planilla de datos** es una **lista de diccionarios**:
- Cada fila de la planilla es un diccionario (la Ficha de Licitación con sus campos).
- Toda la planilla es una lista que almacena estas filas ordenadamente.

Para procesar esto, recorremos la lista con un bucle `for` y, dentro del bucle, accedemos a las claves del diccionario de cada fila.

```python
# Una lista (corchetes) que contiene diccionarios (llaves)
planilla_compras = [
    {"anio": 2024, "monto": 15601131083597},
    {"anio": 2025, "monto": 18430546563769},
]

for fila in planilla_compras:
    # En cada repetición, 'fila' representa a uno de los diccionarios
    anio_actual = fila["anio"]
    monto_actual = fila["monto"]
    print(f"Fila procesada -> Año: {anio_actual} | Gasto: ${monto_actual} CLP")
```

## 6. Comprensión de listas: El tamiz o filtro rápido
Una **comprensión de lista** es una sintaxis compacta de Python que permite crear una nueva lista a partir de otra existente en una sola línea de código. Se usa mucho para **filtrar** o **extraer** información.

### ¿Cómo se lee esta línea?
Supongamos que queremos extraer solo los años donde el monto gastado fue mayor a 13 billones de pesos.

`anios_filtrados = [r["anio"] for r in compras if r["monto"] > 13_000_000_000_000]`

La lectura lógica de izquierda a derecha es:
1.  `[ ... ]`: "Quiero construir una nueva lista..."
2.  `r["anio"]`: "...guardando el año..."
3.  `for r in compras`: "...recorriendo cada registro `r` en la lista `compras`..."
4.  `if r["monto"] > 13_000_000_000_000`: "...pero solamente si el monto de ese registro supera los 13 billones."

```python
compras = [
    {"anio": 2023, "monto": 12483986700171},
    {"anio": 2024, "monto": 15601131083597},
    {"anio": 2025, "monto": 18430546563769},
]

# Filtramos rápidamente usando comprensión de lista
anios_grandes = [r["anio"] for r in compras if r["monto"] > 13_000_000_000_000]
print(anios_grandes)  # Resultado: [2024, 2025]
```

## Errores típicos al empezar (Pistas amigables)
- `IndexError`: Estás pidiendo un número de cajón que no existe en tu lista (ej. pedir `montos[10]` si la lista solo tiene 3 elementos).
- `KeyError`: Estás pidiendo una etiqueta en un diccionario que no está escrita en él (revisa si hay errores de escritura, mayúsculas o tildes en la palabra clave).
- `IndentationError` en bucles: Olvidaste dejar los 4 espacios de sangría debajo de la línea del `for`. Las instrucciones repetidas deben estar indentadas.
- **Confundir corchetes, llaves y paréntesis:**
  - `[ ]` es para **listas** (archivador ordenado).
  - `{ }` es para **diccionarios** (fichas con etiquetas).
  - `( )` es para **tuplas** (datos fijos inmutables).
