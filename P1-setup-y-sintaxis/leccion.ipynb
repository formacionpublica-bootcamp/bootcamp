## 1. Variables: Ponerle etiqueta a una caja
Una **variable** es un nombre que le das a un valor para poder guardarlo y reutilizarlo después. 

**La analogía de la caja de archivo:**
Imagina que tienes una caja física en tu oficina. Le pegas una etiqueta que dice `gasto_2025` y metes dentro un papel con el número `18430546563769`. Cada vez que le pidas a Python que busque en la caja `gasto_2025`, él mirará dentro de la caja y te entregará ese número.

El símbolo `=` no significa "es igual a" en un sentido matemático de ecuación. En programación significa **"asigna"** o **"guarda"** (toma lo de la derecha y guárdalo en la variable de la izquierda).

*   **Reglas para los nombres:** No deben llevar espacios (se usa `guion_bajo`), no pueden empezar con números, no llevan tildes, y Python distingue mayúsculas de minúsculas (`gasto` y `Gasto` son dos cajas distintas).
*   **Comentarios:** Las líneas que empiezan con `#` son notas para ti. Python las ignora por completo al ejecutar el código.

```python
gasto_2025 = 18430546563769
pais = "Chile"

# Le pedimos a Python que nos muestre el contenido de nuestras cajas (variables)
print(gasto_2025)
print(pais)

# Una variable puede cambiar de valor (se queda siempre con la última asignación):
nivel = "bajo"
nivel = "alto"
print(nivel)  # Mostrará "alto"
```

## 2. Tipos de datos: ¿Por qué importa clasificar la información?
En el Estado, no es lo mismo tratar con un **Monto de Gasto** (un número que sumamos) que con un **RUT** o una **Comuna** (texto que no sumamos). Python necesita saber qué tipo de información tiene una variable para saber qué operaciones tiene permitido hacer con ella.

Los tres tipos más básicos son:

| Tipo en Python | Qué guarda | Ejemplo |
| --- | --- | --- |
| `int` (entero) | Números sin decimales | `1867076` |
| `float` (decimal) | Números con coma/punto decimal | `9871357.5` |
| `str` (texto o cadena) | Letras, palabras o números que no se calculan | `"Chile"` o `"2025"` |

**¡Atención!** El texto **siempre** debe ir entre comillas (simples o dobles). Si no pones comillas, Python pensará que estás buscando una variable con ese nombre.

### La diferencia entre sumar números y pegar textos:
Si intentas sumar dos textos, Python no hará matemáticas, sino que los **pegará** (concatenación). 
Para convertir de un tipo a otro, usamos funciones de conversión: `int()`, `float()`, `str()`.

```python
print(type(18430546563769))   # Muestra <class 'int'> (entero)
print(type("Chile"))          # Muestra <class 'str'> (texto)

print("2" + "3")   # Resultado: "23" (concatena o junta los textos)
print(2 + 3)       # Resultado: 5 (suma matemática de enteros)

# Conversión de tipo en acción:
monto_texto = "5000"
monto_numero = int(monto_texto)  # Transforma el texto "5000" al número entero 5000
print(monto_numero + 1000)       # Resultado: 6000 (ahora sí se puede sumar)
```

## 4. Condicionales: Automatizar reglas de negocio
En la administración pública, operamos en base a reglas de negocio. Los condicionales (`if`, `elif`, `else`) le permiten a tu programa tomar decisiones según si estas reglas se cumplen o no.

**La regla de negocio:**
*   **SI (`if`)** el gasto anual es mayor a 10 billones de pesos, clasificar como gasto `"alto"`.
*   **SI NO, PERO SI (`elif`)** el gasto es mayor a 1 billón de pesos, clasificar como gasto `"medio"`.
*   **SI NO SE CUMPLE NINGUNA (`else`)**, clasificar como gasto `"bajo"`.

### La regla de oro en Python: La Indentación
A diferencia de otros lenguajes que usan llaves `{ }` o palabras como `END`, Python usa **la indentación** (los 4 espacios en blanco al inicio de la línea). Esto le indica a Python qué líneas de código se encuentran "dentro" de cada decisión. 

Los comparadores habituales son: `>` (mayor), `<` (menor), `>=` (mayor o igual), `<=` (menor o igual), y **`==` (doble signo igual para comparar si son idénticos)**. *Cuidado:* usar un solo `=` es para guardar datos, no para comparar.

```python
gasto = 18430546563769

if gasto > 10_000_000_000_000:
    nivel = "alto"  # Esta línea está indentada porque depende del 'if'
elif gasto > 1_000_000_000_000:
    nivel = "medio" # Depende del 'elif'
else:
    nivel = "bajo"  # Depende del 'else'

print(nivel)  # Imprimirá "alto"
```

## 6. Trabajar con texto: limpiar y dar formato

Los datos del Estado llegan llenos de texto imperfecto: nombres con espacios de más, mayúsculas inconsistentes ("MINSAL", " ministerio de salud ") o montos guardados como texto. Python trae **métodos** (herramientas pegadas a cada texto) para ordenarlo: se escriben con un punto después del texto.

| Método | Qué hace | Ejemplo |
| --- | --- | --- |
| `.strip()` | Quita espacios al inicio y al final | `"  Salud  ".strip()` → `"Salud"` |
| `.lower()` / `.upper()` | Todo a minúsculas / mayúsculas | `"MINSAL".lower()` → `"minsal"` |
| `.title()` | Primera letra de cada palabra en mayúscula | `"ministerio de salud".title()` → `"Ministerio De Salud"` |
| `.replace(a, b)` | Reemplaza un trozo por otro | `"S.A.G.".replace(".", "")` → `"SAG"` |
| `.split(sep)` | Parte el texto en una lista | `"a,b,c".split(",")` → `["a", "b", "c"]` |

Se pueden **encadenar**: `"  Ministerio DE Salud ".strip().title()` aplica un método tras otro.

### Dar formato a números grandes
Los montos del Estado cuesta leerlos de corrido. Dentro de un f-string, `:,` agrega separador de miles y `.0f` quita los decimales. Ejecuta la celda siguiente para verlo.

```python
# Limpiar texto sucio
nombre_sucio = "  ministerio DE salud  "
print(nombre_sucio.strip())          # quita los espacios de los extremos
print(nombre_sucio.strip().title())  # "Ministerio De Salud" (encadenado)
print("MINSAL".lower())              # "minsal"
print("S.A.G.".replace(".", ""))     # "SAG"

# Formatear un monto grande para que sea legible
gasto = 18430546563769
print(f"Gasto 2025: ${gasto:,.0f} CLP")   # $18,430,546,563,769
# (en Chile el separador de miles es el punto; aquí Python usa coma. Lo ajustaremos más adelante.)
```

## 7. Una pausa clave: ¿Qué es una función (`def` y `return`)?
Para corregir automáticamente tus ejercicios en este bootcamp, usamos celdas de chequeo que prueban tu código con varios montos diferentes para asegurarse de que tu lógica sea correcta y no se caiga. Para lograr esto, tu código debe estructurarse dentro de una **función**.

**La analogía del trámite o plantilla de oficio:**
Una función es como un trámite administrativo predefinido o una plantilla de oficio:
1.  **`def` (definir):** Es la orden para crear un trámite o máquina con un nombre claro, por ejemplo: `def clasificar_gasto(monto):`.
2.  **Parámetro (lo que va entre paréntesis):** Es el "formulario de entrada" o dato inicial que la función necesita para poder trabajar (en este caso, un `monto`).
3.  **Cuerpo:** El conjunto de instrucciones (indentadas) que procesan la información de entrada.
4.  **`return` (devolver):** Es el producto final firmado que la máquina le entrega de vuelta a quien le solicitó el trámite.

### Diferencia fundamental: `print` vs `return`
*   `print()` es solo una pantalla: "dibuja" letras en tu pantalla de Colab para que el ojo humano las lea, pero el programa no puede reutilizar ese valor.
*   `return` entrega un valor real al sistema. La celda de chequeo automático necesita que la función **devuelva (`return`)** el resultado para validar si está correcto.

### ¿Qué significa `pass`?
Es simplemente una palabra de relleno en Python que le dice "no hagas nada". En los ejercicios la colocamos para que el notebook no se rompa antes de que tú escribas tu solución. **Debes borrar el `pass` e ingresar tu código allí.**

## Errores típicos al empezar (¡No te asustes!)
Si ejecutas una celda y ves letras rojas, lee la última línea del mensaje de error, es una pista amigable:
- `SyntaxError` (Error de sintaxis): Te faltó cerrar unas comillas, unos paréntesis, o te faltaron los `:` obligatorios al final de un `if` o `def`.
- `NameError` (Error de nombre): Estás llamando a una variable que no has creado todavía, o escribiste mal su nombre (recuerda que `Gasto` y `gasto` no son lo mismo).
- `IndentationError` (Error de indentación): Olvidaste dar los 4 espacios de sangría debajo de un `if` o un `def`.
- El texto te muestra literalmente `{anio}`: Olvidaste colocar la letra `f` antes de iniciar las comillas del texto.
