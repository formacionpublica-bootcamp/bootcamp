## 1. Funciones: El trámite administrativo o plantilla de oficio
Una **función** es un bloque de código con nombre que realiza una tarea específica, permitiendo reutilizar la misma lógica muchas veces sin tener que volver a escribirla.

**La analogía del trámite o la plantilla de oficio:**
Imagina que una función es una "máquina de trámites" del servicio público o una plantilla de oficio automatizada:
1.  **`def nombre_funcion(parametros):`** Es el acto de definir el trámite, indicando qué nombre tiene y qué "documentos de entrada" (parámetros) requiere obligatoriamente para poder procesar.
2.  **El cuerpo (indentado):** El conjunto de pasos o reglas de negocio que se aplican a los documentos ingresados.
3.  **`return resultado`:** Es la entrega del producto final firmado que se devuelve al solicitante.

### El parámetro por defecto (El campo pre-rellenado)
Imagina un formulario del Estado donde si no especificas un campo, el sistema asume automáticamente un valor por defecto. Por ejemplo, en una función de conversión financiera, si no especificas el divisor, asumiremos por defecto que quieres dividir por un millón (`1_000_000`) para ver cifras resumidas.

```python
# Definimos una función simple
def gasto_promedio(monto, ordenes):
    return monto / ordenes  # Devuelve el resultado del cálculo

# Llamamos a la función con datos reales de 2025
print(gasto_promedio(18430546563769, 1867076))

print("------------------")

# Función con parámetro por defecto (si no se entrega divisor, divide por 1 millón)
def en_millones(monto, divisor=1_000_000):
    return monto / divisor

print(en_millones(18430546563769))          # Usa el divisor por defecto (1.000.000)
print(en_millones(18430546563769, 1000))    # Usa otro divisor entregado a mano (1.000)
```

## 2. Módulos: La caja de herramientas de otra oficina
Python viene con una gran cantidad de funcionalidades preconstruidas organizadas en **módulos** (o librerías). En lugar de escribir fórmulas complejas desde cero, podemos "importar" estas herramientas hechas por otros desarrolladores utilizando la instrucción `import`.

**La analogía de las herramientas del servicio estándar:**
Imagina que para calcular estadísticas básicas de tu departamento no tienes que escribir la fórmula matemática del promedio. Simplemente tomas prestada la calculadora oficial llamada `statistics` que ya sabe calcular promedios (`mean`) y medianas (`median`).

## 3. Leer archivos externos: El cajón del archivador
Para analizar datos públicos de verdad, necesitamos abrir archivos que están guardados en el disco duro de la computadora.

**La analogía del cajón del archivador:**
1.  **Abrir el cajón:** La instrucción `open(ruta)` abre una conexión al archivo.
2.  **Leer el documento:** Extraemos el contenido del archivo.
3.  **Cerrar el cajón con llave:** Es de vital importancia cerrar el archivo una vez que terminamos de usarlo. Si no lo cerramos, el archivo queda bloqueado en la memoria de la computadora, consumiendo recursos y arriesgándose a romperse.

### El asistente ordenado: El bloque `with`
Para evitar que olvides cerrar el archivo por descuido, en Python usamos la estructura `with open(ruta) as alias:`. 
Funciona como contratar a un asistente de oficina muy ordenado: él abre el archivador, te pasa los datos para que trabajes dentro del bloque indentado, y **se asegura de cerrar el cajón con llave inmediatamente** al salir del bloque, ocurra lo que ocurra.

*(Nota: Para esta lección, el archivo `compras.csv` ya viene guardado estáticamente en esta carpeta de tu proyecto. Fuente de los datos: MercadoPúblico / ChileCompra en datos.gob.cl)*

Ahora leeremos el archivo completo y lo mostraremos en pantalla usando a nuestro asistente ordenado `with`:

## 4. Leer un CSV como registros (Planilla Excel)
Un archivo **CSV** (Comma-Separated Values) es simplemente una planilla de cálculo básica guardada como texto, donde cada columna se separa por una coma `,`.

El módulo nativo `csv` de Python cuenta con una herramienta llamada `csv.DictReader` que lee este archivo y **transforma automáticamente cada fila en una Ficha de Licitación (diccionario)**, usando la primera línea del archivo (el encabezado) como etiquetas de las claves.

### ⚠️ La trampa del texto (`str`) al leer archivos:
Cuando Python lee un archivo de texto, **toda la información ingresa como texto (`str`)**, incluso si ves dígitos numéricos. 
Si lees el monto `"9835060128510"` de una fila, para Python son solo caracteres dibujados. Si intentas sumarlos, los pegará en lugar de sumarlos. **Debes convertir explícitamente el valor a entero con `int()`** antes de realizar cálculos matemáticos.

```python
import csv

# Leemos el CSV fila por fila como diccionarios
with open("compras.csv", encoding="utf-8") as f:
    for fila in csv.DictReader(f):
        # Cada 'fila' es un diccionario: clave = columna, valor = celda (como texto)
        print(f"Año: {fila['anio']} | Gasto en texto: {fila['monto']}")

print("------------------")

# Ejemplo: Sumar la columna 'monto' convirtiéndola a número entero
total_gasto = 0
with open("compras.csv", encoding="utf-8") as f:
    for fila in csv.DictReader(f):
        # Transformamos el texto de la columna 'monto' a un número entero real
        monto_numerico = int(fila["monto"])
        total_gasto = total_gasto + monto_numerico

print(f"El total acumulado real es: ${total_gasto} CLP")
```
